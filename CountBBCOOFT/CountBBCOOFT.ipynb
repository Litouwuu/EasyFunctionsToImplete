```python
import pandas as pd
import numpy as np
```

```python
Rawdata = pd.read_excel('COUNTOFD_BBCO_OFT_DATABASE.xlsx', header = 0)
```

```python
InUse = ['SKUCode', 'ReleaseDate', 'ActualDeliveryDate','OrderNum','DeliveryOrderVolumeHL']
df = Rawdata[InUse]
```

```python
group_Vol = df.groupby('OrderNum').agg({'DeliveryOrderVolumeHL':np.sum})
group_Vol = group_Vol.rename(columns={'DeliveryOrderVolumeHL':'OrderVol'})
group_Vol
```

```python
merge = pd.merge(df,group_Vol,how = 'left',on='OrderNum')
merge['part_OFT'] = (merge['ActualDeliveryDate'] - merge['ReleaseDate']).map(lambda x: x.days)*(merge['DeliveryOrderVolumeHL']/merge['OrderVol'])
merge.head(10)
```

```python
Final_merge = merge.groupby('OrderNum')['part_OFT'].sum()
Final_merge.mean()
```
