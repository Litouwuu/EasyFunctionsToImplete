### Question 1 Energy

```python
# Written by myself
Energy = pd.read_excel('assets/Energy Indicators.xls',sheet_name='Energy',
                       header=17,skipfooter=38,usecols= [2,3,4,5])
Energy = Energy.fillna(np.nan).reset_index()
# Energy['Energy Supply']= Energy['Petajoules'].map(lambda x : x if type(x) == int else np.nan)
# Energy['Energy Supply per Capita'] = Energy['Gigajoules'].map(lambda x : x if type(x) == int else np.nan)
# Energy.drop(columns=['Petajoules','Gigajoules'])
# Energy = Energy[['Energy Supply','Energy Supply per Capita','%']]# 
# 'Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable
Energy.rename(columns={'index':'Country','Petajoules':'Energy Supply',
                       'Gigajoules':'Energy Supply per Capita','%':'% Renewable'},inplace = True)
Energy['Energy Supply']=Energy['Energy Supply'].map(lambda x : x if type(x) == int else np.nan)*1000000
Energy['Energy Supply per Capita'] = Energy['Energy Supply per Capita'].map(lambda x : x if type(x) == int else np.nan)

Energy.replace({"Republic of Korea": "South Korea", "United States of America": "United States", 
                "United Kingdom of Great Britain and Northern Ireland": "United Kingdom", 
                "China, Hong Kong Special Administrative Region": "Hong Kong"},inplace = True)
Energy['Country'] = Energy['Country'].str.replace(r'\(.*\)','')
Energy['Country'] = Energy['Country'].str.replace(r'\d*','')
```
