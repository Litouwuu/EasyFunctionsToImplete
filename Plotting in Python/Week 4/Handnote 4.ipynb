# DataFrame.plot

```python
np.random.seed(123)
df = pd.DataFrame({'A':np.random.randn(365).cumsum(0),
                  'B':np.random.randn(365).cumsum(0)+20,
                  'C':np.random.randn(365).cumsum(0)-20},
                 index = pd.date_range('1/1/2021',periods=365))
df.head()
```

```python
df.plot()
```

```python
df.plot('A','B',kind='scatter')
```

```python
ax = df.plot.scatter('A','C',c=df['B'],s =df['B'],colormap='viridis')
ax.set_aspect('equal')
```

```python
df.plot.box()
```

# Seaborn

```python
import seaborn as sns
```

```python
np.random.seed(1234)

v1 = pd.Series(np.random.normal(0,10,1000), name = 'v1')
v2 = pd.Series(2*v1 + np.random.normal(60,15,1000),name='v2')
```

```python
plt.figure()
plt.hist(v1,alpha=0.7,bins=np.arange(-50,150,5),label='v1')
plt.hist(v2,alpha=0.7,bins=np.arange(-50,150,5),label='v2')
plt.legend()
```

```python
plt.figure()
plt.hist([v1,v2],alpha = 0.5,histtype='barstacked',normed=True)
v3= np.concatenate((v1,v2))
sns.kdeplot(v3)
```

```python
plt.figure()
sns.distplot(v3,hist_kws={'color':'Teal'})
```

```python
sns.jointplot(v1,v2,alpha=0.3)
```

```python
sns.set_style('white')
sns.jointplot(v1,v2,kind='kde',space=0)
```
